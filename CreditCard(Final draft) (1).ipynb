# Spliting Data Into Features and Target

```python
X=data.drop('Class', axis=1)
y=pd.Series(data.Class)
```

Spliting Data Into Train/Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

# Preprocessing & Dimentionality Reduction

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

Model Support Vector Machine Training

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
n_components=10
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('pca', PCA(n_components=n_components)),
    ('classifiers', SVC())
])
```

```python
pipeline.fit(X_train, y_train)
```

# Testing Model

```python
y_pred=pipeline.predict(X_test)
y_pred=pd.DataFrame(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_pred, y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix1= confusion_matrix(y_test, y_pred)
print(confusion_matrix1)
```

# Other Accuracy Measures

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

# Tuning Hyperparameters

```python
scaler= StandardScaler()
scaled_data=scaler.fit_transform(X)
pca=PCA(n_components=10)
most_relevent_features=pca.fit_transform(scaled_data)
```

```python
from sklearn.model_selection import train_test_split

Xscaled_train, Xscaled_test, y_train, y_test = train_test_split(most_relevent_features, y, test_size=0.30, random_state=1)
```

```python
model=SVC()
from sklearn.model_selection import GridSearchCV
para_grid= {
    'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [0.1, 1.0, 10]
}
grid_search= GridSearchCV(model, param_grid=para_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(Xscaled_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred1=best_model.predict(Xscaled_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred1)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix01= confusion_matrix(y_test, y_pred1)
print(confusion_matrix01)
```

# Other Accuracy Measures

```python
TP = confusion_matrix01[1][1]
FP = confusion_matrix01[0][1]
FN = confusion_matrix01[1][0]

# Calculate precision
precision = TP / (TP + FP)

# Calculate recall
recall = TP / (TP + FN)

# Calculate F1 score
f1_score = 2 * (precision * recall) / (precision + recall)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1_score}")
```

# Hard Voting(Ensemble)

```python
from sklearn.ensemble import VotingClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB


clf_01 = LogisticRegression()
clf_02 = SVC(probability=False)  
clf_03 = DecisionTreeClassifier()
clf_04 =SGDClassifier()
clf_05 =GaussianNB()

# Training Individual models 
clf_01.fit(Xscaled_train, y_train)
clf_02.fit(Xscaled_train, y_train)
clf_03.fit(Xscaled_train, y_train)
clf_04.fit(Xscaled_train, y_train)
clf_05.fit(Xscaled_train, y_train)

# Predictions from Individual models 
pred01= clf_01.predict(Xscaled_test)
pred02= clf_02.predict(Xscaled_test)
pred03= clf_03.predict(Xscaled_test)
pred04= clf_04.predict(Xscaled_test)
pred05= clf_05.predict(Xscaled_test)
# Traing Ensemble 
ensemble = VotingClassifier(estimators=[('lr', clf_01), ('svm', clf_02), ('dt', clf_03),('sgd', clf_04), ('gnb', clf_05)], voting ='hard')

ensemble.fit(Xscaled_train, y_train)

predictions = ensemble.predict(Xscaled_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
accuracy01=accuracy_score(y_test, pred01)
accuracy02= accuracy_score(y_test, pred02)
accuracy03= accuracy_score(y_test, pred03)
accuracy04= accuracy_score(y_test, pred04)
accuracy05= accuracy_score(y_test, pred05)
accuracyHV= accuracy_score(y_test, predictions)

print("Accuracy of LogesticRegression", accuracy01)
print("Accuracy of SVM", accuracy02)
print("Accuracy of DecisionTree", accuracy03)
print("Accuracy of StochasticGradient", accuracy04)
print("Accuracy of  Naive Bayes Classifier", accuracy05)
print("Accuracy of HardVoting", accuracyHV)
```
